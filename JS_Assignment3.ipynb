function addStudent(id, firstName, lastName, age, grade) {
  const exists = students.some(student => student.id === id)
  if (exists) {
    console.log(`ID ${id} already exists. Please use another ID.`);

function updateStudentInfo(id, newinfo) {
  const index = students.findIndex(student => student.id === id);
  if (index === -1) {
    console.log(`Student ID ${id} does not exists`)
  }
  else {
    students[index] = { ...students[index], ...newinfo };
    console.log(`Student with ID ${id} updated successfully.`);
    return students;
  }

function deleteStudent(id) {
  const index = students.findIndex(student => student.id === id);
  if (index === -1) {
    console.log(`Student ID ${id} does not exists`)
  }
  else {
    students.splice(index, 1);// 1 item from that index will be removed here

}

// 5. Find Students by Grade: Create a function to find all students who have a specific grade.

function studentsByGrade(grade) {
  const matchedStudents = students.filter(student => student.grade === grade);

  if (matchedStudents.length === 0) {
    console.log(`The Grade ${grade} is not found`);
    return;
  }

  for (let student of matchedStudents) {
    console.log(`ID: ${student.id}, Name: ${student.firstName}`);
  }
}

// 6. Calculate Average Age: Create a function to calculate the average age of all students using array method

function avgAge() {
  if (students.length === 0) {
    console.log('No students are available');
    return;
  }
  let sum = 0
  for (let student of students) {
    sum += student.age;
  }
  let average = sum / students.length;
  console.log(`Average age of students is ${average}`);
  return average;
}
